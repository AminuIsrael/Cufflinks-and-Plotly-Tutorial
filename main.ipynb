```python
#importing modules
import plotly
import cufflinks as cf
cf.go_offline()  #This lets us make our plots offline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Collecting Data

```python
#Most of the data collected was via web scraping and from the Nigeria Bureau of Statistics
```

```python
Data0 = pd.read_csv("Consumption.csv") #Data on Nigeria electricity consumption (1971-2014)
Data1 = pd.read_csv("Jamb_no_of_applicants(2017-2018)") #jamb 2017-2018 data
Data2 = pd.read_csv("DANGCEM STOCK") #Dangote cement stock from march to july
```

## Line Plots

#### DANGCEM line plot

```python
Data2.iplot(x="Date",y=["Open","High","Low","Change","Price"],title="DANGOTE CEMENT STOCK(MAR-APR)",
            theme="solar")
```

### JAMB Data Line plot

```python
Data1.iloc[0:37,:].iplot(x="State",y=["Female_2017","Male_2017"],theme='henanigans',
                         title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FOR 2017")
```

```python
Data1.iloc[0:37,:].iplot(x="State",y=["Females_2018","Males_2018"],
                         theme='henanigans',title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FOR 2018")
```

```python
Data1.iloc[0:37,:].iplot(x="State",y=["Female_2017","Females_2018","Male_2017","Males_2018"],
                         theme='polar',title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FOR 2018")
```

#### Nigeria Electricity comsumtion line plot

```python
Data0.iplot(x="Year",y="Value(Kwh)",theme='pearl',
            title="Nigeria Electricity Consumption(Kwh) 1971-2014",yTitle="Value(kwh)",xTitle="Year")
```

```python
cf.getThemes() #You could use these other themes too :) ...
#you could set a default theme using the syntax cf.set_config_file(theme='pearl')
```

### Bar Plots

### Jamb Data Bar plot

```python
Data1.iloc[0:37,:].iplot(kind="bar",x="State",y=["Female_2017"],theme='polar',
                         title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FOR FEMALES 2017")
```

```python
Data1.iloc[0:37,:].iplot(kind="bar",x="State",y=["Male_2017","Female_2017"],theme='polar',
                         title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FOR 2017")
```

```python
Data1.iloc[0:37,:].iplot(kind="bar",x="State",y=["Females_2018","Males_2018"],theme='polar',
                         title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FOR 2018")
```

```python
#You can change the orientation by using the orientation = "h" or "v" syntax
Data1.iloc[0:37,:].iplot(kind="bar",x="State",y=["Females_2018","Males_2018"],theme='polar',
                         title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FOR 2018",orientation='h')
```

```python
Data1.iloc[0:37,:].iplot(kind="bar",x="State",y=["Male_2017","Female_2017","Females_2018","Males_2018"],theme='space',
                         title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FROM 2017-2018")
#This plot doesnt look good for presentation, instread we could do something...
```

```python
#You could plot a group bar chart instead by setting barmode= "stack"
Data1.iloc[0:37,:].iplot(kind="bar",barmode="stack",x="State",y=["Male_2017","Female_2017","Females_2018","Males_2018"],theme='space',
                         title="JAMB APPLICANTS STATISTICS BY STATE AND GENDER FROM 2017-2018")
#This plot carries every details about the applicants and year
#by double clicking on one of the legends youll see the individual plots.
```
